```python

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set(style="whitegrid")
```

```python
from google.colab import files
uploaded = files.upload()

df = pd.read_csv("diabetes_cleaned.csv")
df.head()
```

 Distribution Plot (Histogram)



```python
plt.figure(figsize=(8,5))
sns.histplot(df['Glucose'], kde=True)
plt.title("Distribution of Glucose Levels")
plt.xlabel("Glucose Level")
plt.ylabel("Frequency")
plt.show()
```

Scatter Plot (Relationship)

```python
plt.figure(figsize=(6,5))
sns.scatterplot(x='BMI', y='Glucose', hue='Outcome', data=df)
plt.title("Relationship Between BMI and Glucose by Diabetes Outcome")
plt.xlabel("BMI")
plt.ylabel("Glucose")
plt.show()
```

 Bar Chart (Comparison)

```python
plt.figure(figsize=(6,5))
sns.countplot(x='Outcome', data=df)
plt.title("Distribution of Diabetes Cases")
plt.xlabel("Outcome (0 = No Diabetes, 1 = Diabetes)")
plt.ylabel("Count")
plt.show()
```

Line Plot (Trend)

```python

age_diabetes = df.groupby('Age')['Outcome'].mean()

plt.figure(figsize=(8,5))
age_diabetes.plot()
plt.title("Trend of Diabetes Probability by Age")
plt.xlabel("Age")
plt.ylabel("Average Diabetes Rate")
plt.show()
```

Part 2: Feature Engineering

BMI Category Feature

```python
df['BMI_Category'] = pd.cut(
    df['BMI'],
    bins=[0, 18.5, 25, 30, 100],
    labels=['Underweight', 'Normal', 'Overweight', 'Obese']
)
```

Glucose Category Feature


```python

df['Glucose_Category'] = pd.cut(
    df['Glucose'],
    bins=[0, 99, 125, 300],
    labels=['Normal', 'Prediabetes', 'Diabetes'])
```
